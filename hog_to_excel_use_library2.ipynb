```python
def hog_to_list(image_file):
    img = cv2.imread(image_file)
    if(len(img) <len(img[0])):
        img = cv2.rotate(img, cv2.ROTATE_90_COUNTERCLOCKWISE)


    img =cv2.resize(img, (64,128))

    #print(img.shape[0:2])



    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    #cv2.imshow('img',img)
    MC= False
    # hogfv,hog_image = hog(img, orientations=8, pixels_per_cell=(16, 16),
    #                 cells_per_block=(2, 2), visualize=True, multichannel=True)
    # hogfv,hog_image = hog(img,orientations=9,pixels_per_cell = (8,8),cells_per_block =(2,2),
    #                       visualize=True, multichannel=MC)
    hogfv,hog_image = hog(img,orientations=9,pixels_per_cell = (8,8),cells_per_block =(2,2),
                          visualize=True, multichannel=MC)
    hog_image_rescaled = exposure.rescale_intensity(hog_image,in_range=(0,5))
    hog_feature = hogfv.tolist()
    return hog_feature



```
