```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def grad_desc_momentum(x0, f, df, lr=1e-2, eps=1e-2, alpha=1e-2, show=True):
  x_list = []
  y_list = []
  xOld = x0
  xNew = xOld +2*eps
  dw = 0
  i = 0
  while abs(xNew -  xOld) > eps:
    xOld = xNew
    dw = lr * df(xNew) + alpha*dw
    print(dw)
    xNew -= dw
    if show:
      x_list.append(xNew)
      y_list.append(f(xNew))
      #plt.plot(xNew, f(xNew), '.r')
      #plt.text(xNew, f(xNew), str(i))
      #plt.draw()
      #plt.pause(0.2)
      #print(i)
      i += 1
  return xNew,x_list,y_list,i
```

```python
#normal gd
def f(x):
  return 2*x**4+5*x**3-30*x**2+x-1

def df(x):
  return 8*x**3+15*x**2-60*x+1

x = np.arange(-5,5.1,0.1)
y = f(x)
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')

local_min = grad_desc_momentum(5, f, df, lr=1e-3, eps= 1e-3,alpha = 0)
plt.plot(local_min[1], local_min[2], '.r')

exact_min = -3.83775470916525
print('Gradient Descent\'s Local minimum occurs at' ,local_min[0])
print('Exact minimum occurs at ',exact_min)
print('Erro rate is ',abs(local_min[0] - exact_min)/ exact_min * 100, '%')
print('use ',local_min[3],' round')
plt.show()
```

```python
def grad_desc_projectile(x0, f, df, lr=1e-2, eps=1e-2, alpha=1e-2, show=True):
  x_list = []
  y_list = []
  xOld = x0
  xNew = xOld +2*eps
  dw = 0
  i = 0
  while abs(xNew -  xOld) > eps:
    xOld = xNew
    dw = lr * df(xNew) + alpha*dw*1/np.sqrt(2)
    print(dw)
    xNew -= dw
    if show:
      x_list.append(xNew)
      y_list.append(f(xNew))
      #plt.plot(xNew, f(xNew), '.r')
      #plt.text(xNew, f(xNew), str(i))
      #plt.draw()
      #plt.pause(0.2)
      #print(i)
      i += 1
  return xNew,x_list,y_list,i
```

```python
def grad_desc(x0, f, df, lr=1e-2,eps=1e-2,alpha= 1e-2, show=True):
  x_list = []
  y_list = []
  xOld = x0
  xNew = xOld + 2*eps
  dw = 0
  i = 0
  while abs(xNew - xOld) > eps :
    xOld = xNew
    dw  = lr * df(xNew) + alpha * dw*1/2
    xNew -= dw
    if show:
      x_list.append(xNew)
      y_list.append(f(xNew))
      i += 1
  return xNew,x_list,y_list,i

def f(x):
  return x**4-3*x**3+2
def df(x):
  return 4*x**3-9*x**2

x = np.arange(2,2.41,0.01)
y = f(x)
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')


local_min = grad_desc(2,f,df,alpha=1)
plt.plot(local_min[1],local_min[2],'.r')

exact_min = 9/4
print('Gradient Descent\'s Local minimum occurs at ',local_min[0])  
print('Exact minimum occurs at ',exact_min)
print('Error rate is ',abs((local_min[0] - exact_min)/exact_min * 100),'%')    
print('use ',local_min[3],' round')
```
