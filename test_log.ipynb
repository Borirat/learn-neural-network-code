```python
def y_relu(x):
  x_new = np.zeros(len(x))
  for i in range(len(x)):
    if(x[i]>0):
      x_new[i] = x[i]
    else:
      x_new[i] = 0
  return x_new

plt.plot(x_in,(y_relu(x_in)),'.r')
plt.show()
```

```python
def new_y_relu(x):
  x_new = np.zeros(len(x))
  for i in range(len(x)):
    if(x[i]>0):
      x_new[i] = 1
    else:
      x_new[i] = 0
  return x_new

plt.plot(x_in,(new_y_relu(x_in)),'.r')
plt.show()
```

```python
x_in3 = np.arange(-10.0,10.0,0.1)
def tan_with_pi(x):
  x_new = np.zeros(x.shape)
  for i in range(len(x)):
    x_new[i] =  2*np.arctan(x[i])/np.pi
  return x_new

plt.plot(x_in3,tan_with_pi(x_in3),'.r')
plt.show()
```

```python
def de_tan_with_pi(x):
  x_new = np.zeros(x.shape)
  for i in range(len(x)):
    x_new[i] =  2/(np.pi*(1+x[i]**2))
  return x_new

plt.plot(x_in3,de_tan_with_pi(x_in3),'.r')
plt.show()
```

```python
def x_pow2_divide_by_x_pow2_plus_1(x):
  x_new = np.zeros(x.shape)
  for i in range(len(x)):
    x_new[i] =  x[i]**2/(x[i]**2+1)
  return x_new

plt.plot(x_in3,x_pow2_divide_by_x_pow2_plus_1(x_in3),'.r')
plt.show()
```

```python
def diff_x_divide_pow2_by_x_pow2_plus_1(x):
  x_new = np.zeros(x.shape)
  for i in range(len(x)):
    x_new[i] =  (2*x[i])/((x[i]**2+1)**2)
  return x_new

plt.plot(x_in3,diff_x_divide_pow2_by_x_pow2_plus_1(x_in3),'.r')
plt.show()
```

```python
#2.718281828
x_in6 = np.arange(-100.0,100.0,0.1)
def log_e(x):
  x_new = np.zeros(x.shape)
  for i in range(len(x)):
    if(x[i]>=0):
      x_new[i] =  (np.log(x[i])/24)+0.7
    else:
       x_new[i] =  (0- (np.log(abs(x[i]))/24))+0.2
  return x_new

plt.plot(x_in6,log_e(x_in6),'.r')
plt.show()
```

```python
x_in6 = np.arange(-100.0,100.0,0.1)
def e_x(x):
  x_new = np.zeros(x.shape)
  for i in range(len(x)):
      x_new[i] =  exp(1)*x[i]**2/36000
    
  return x_new

plt.plot(x_in6,e_x(x_in6),'.r')
plt.show()
max(e_x(x_in6))
```
