```python
weight_list = []
for i in range(40,121):
    if(i<46):
        for r in range(3):
            weight_list.append(i)
    elif(i>45 and i < 86):
        for r in range(5):
            weight_list.append(i)
    else:
        weight_list.append(i)
hight_list = []
for i in range(120,191):
    if(i<135):
        hight_list.append(i)
    elif(i>134 and i < 151):
        for r in range(3):
            hight_list.append(i)
    elif(i>150 and i < 171):
        for r in range(5):
            hight_list.append(i)
    else:
        for r in range(2):
            hight_list.append(i)
print(weight_list)
print(hight_list)
```

```python
number_data = 200
p_weight = np.random.choice(weight_list,number_data)
hight = np.random.choice(hight_list,number_data)
# print(weight)
# print(hight)
thin_weight = []
thin_hight = []
fat_weight = []
fat_hight = []
target = []
for i in range(0,number_data):
#     print(weight[i]," ",hight[i]," ",weight[i]/((hight[i]/100)**2))
    if(p_weight[i]/((hight[i]/100)**2)<22.9):
        thin_weight.append(p_weight[i])
        thin_hight.append(hight[i])
        target.append(0) # 0 คือ ผอม
    else:
        fat_weight.append(p_weight[i])
        fat_hight.append(hight[i])
        target.append(1) # 1 คือ อ้วน
```

```python
def train_weight(input_data,weight,bias,target,result):
    while(result != target):
        error = target-result
        weight = weight+np.dot(error,input_data)
        bias = bias+error*5
        result = hard_limit_function(np.dot(weight,np.transpose(input_data))+bias)
        print("in train mode : error = ",error,"weight = ",weight," bias = ",bias," result = ",result)
    return weight,bias,result
```

```python
def classify(input_data,weight,bias,target):
    result = perceptron(input_data,weight,bias)
    if(result==target):
        return True
    else:
        return False
        
```

```python
def train_until_weight_suit_for_classify(p_weight,hight,weight,bias,target):
    round_for_train = 0
    check_true = 0
    while (check_true <198 and round_for_train<200):
        round_for_train+=1
        check_true = 0
        for i in range(len(p_weight)):
            result = perceptron(np.array([p_weight[i]/100,hight[i]/100]),weight,bias)
            if(result==target[i]):
                check_true +=1
            else:
                weight,bias,result = train_weight(np.array([p_weight[i]/100,hight[i]/100]),weight,bias,target[i],result)
                check_true +=0
    print("Now,Weight are OK! and use round for train = ",round_for_train)
    return weight,bias
```
