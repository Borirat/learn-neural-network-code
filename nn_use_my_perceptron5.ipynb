**CREATE WEIGHT**

```python
WIL_R = [] # มีทั้งหมด 32 ส่วน แต่ละส่วน ประกอบด้วย 2 row 3 column แต่ต้องสลับ
for i in range(32):
    WIL_R.append(np.random.rand(3,2).astype(np.float) )
WIL_R
```

```python
WLT_R = [] # มีทั้งหมด 32 ส่วน  แต่ละส่วน ประกอบด้วย 3 row 4 column แต่ต้องสลับ
for i in range(32):
    WLT_R.append(np.random.rand(4,3).astype(np.float) )
WLT_R
```

**CREATE BIAS**

```python
b_for_WIL_R = np.ones(96).astype(np.float) # 96 จาก 32 จุด โดยแต่ละจุดมี ไป 3 tf_L
b_for_WIL_R
```

```python
b_for_WLT_R  = np.ones((32,4)).astype(np.float) # 128 เนื่องจากว่า 32 จุด แต่ละจุดไปยัง 4 tf_T
b_for_WLT_R
```

**PERCEPTRON**

```python
def perceptron(input_data,weight,bias):
    # print(np.dot(input_data,np.transpose(weight))+bias)
    return np.dot(input_data,np.transpose(weight))+bias
```

**TRANSFER FUNCTION**

```python
def tf_L1(result_of_summation):
    # print("result_of_summation = ",result_of_summation)
    if(result_of_summation>=0.8):
        return 1
    else:
        return 0
```

```python
def tf_L2(result_of_summation):
    # print("result_of_summation = ",result_of_summation)
    if(result_of_summation>=0.5):
        return 1
    else:
        return 0
```

```python
def tf_L3(result_of_summation):
    # print("result_of_summation = ",result_of_summation)
    if(result_of_summation>=0.3):
        return 1
    else:
        return 0
```

```python
def tf_L2T(result_of_summation):
    #print("result_of_summation = ",result_of_summation)
    for i in range(len(result_of_summation)):
      if(result_of_summation[i]>=2):
          result_of_summation[i] = 1
      else:
          result_of_summation[i] = 0
    return result_of_summation
```

**TRAIN WEIGHT**

```python
def tr_L(delta_weight_trans,WIL_R,b_for_WIL_R,in_r,w_position,bias_postion):
    #print("w = ",WIL_R[in_r])
    delta_weight_in = (sum(delta_weight_trans)*WIL_R[in_r][w_position])/sum(WIL_R[in_r][w_position])
    WIL_R[in_r][w_position] = WIL_R[in_r][w_position]+delta_weight_in
    b_for_WIL_R[bias_postion] = b_for_WIL_R[bias_postion]+sum(delta_weight_trans)*0.5
    #print('bias delta = ',sum(delta_weight_trans))
    return WIL_R,b_for_WIL_R
    #print("delta = ",delta_weight_in)
    # #error = sum(delta_weight_trans))
    # diff_error = delta_weight_trans*transfer_line
    # print(transfer_line)
    # delta_weight = (diff_error*WIL_R[in_r]) / sum(WIL_R[in_r])*0.1
    # print(delta_weight)
    
```

```python
def tr_L2T(target,result,WLT_R,b_for_WLT_R,in_r):
    target_array = 0
    if(target == 1):
          target_array = np.array([1,0,0,0]).astype(np.float)
    elif(target == 2):
          target_array = np.array([0,1,0,0]).astype(np.float)
    elif(target == 3):
          target_array = np.array([0,0,1,0]).astype(np.float)
    elif(target == 4):
          target_array = np.array([0,0,0,1]).astype(np.float)

    error = target_array-result
    
    delta_weight = []
    for i in range(len(WLT_R[in_r])):
        delta_weight.append(error[i]*WLT_R[in_r][i] / sum(WLT_R[in_r]))
    
    # print('dleta weight ',delta_weight)
    WLT_R[in_r] = WLT_R[in_r]+delta_weight
    b_for_WLT_R[in_r] = b_for_WLT_R[in_r]+sum(error)*0.5
    #print(WLT_R[in_r])
    #return WLT_R,b_for_WLT_R,error,delta_weight
    return WLT_R,b_for_WLT_R,delta_weight
```

**LEARN**

```python
def lr_r_bins(input_r,bins_order,WIL_R,WLT_R,b_for_WIL_R,b_for_WLT_R,learn_round):
  round_train = 0
  check_true = 0

  while (True and round_train<learn_round):
    round_train+=1
    check_true = 0

      #หาค่า
      ###################
    for pic_input_r in range(len(input_r)): # จำนวนของภาพในแชลเนลสีแดง
        for in_r in range(len(input_r[0])): # in_r คือ จุด
          #L1
          percept_L1 =  perceptron(np.array([bins_order[in_r],input_r[pic_input_r][in_r]]),WIL_R[in_r][0],b_for_WIL_R[in_r*3]) #in_r*3 มาจากวนให้ถึง 96
          transfer_L1 =  tf_L1(percept_L1)

          #L2
          percept_L2 =  perceptron(np.array([bins_order[in_r],input_r[pic_input_r][in_r]]),WIL_R[in_r][1],b_for_WIL_R[in_r*3+1])
          transfer_L2 =  tf_L2(percept_L2)

          #L3
          percept_L3 =  perceptron(np.array([bins_order[in_r],input_r[pic_input_r][in_r]]),WIL_R[in_r][2],b_for_WIL_R[in_r*3+2])
          transfer_L3 =  tf_L1(percept_L3)

          #type
          percept_type  = perceptron(np.array([transfer_L1,transfer_L2,transfer_L3]),WLT_R[in_r],b_for_WLT_R[in_r])
          transfer_type  = tf_L2T(percept_type)
          print("result for type = ",transfer_type)

          #เช็ค error แล้วค่อยดำเนินการอัปเดต
          target_array = 0
          if(target[pic_input_r] == 1):
                target_array = np.array([1,0,0,0]).astype(np.float)
          elif(target[pic_input_r] == 2):
                target_array = np.array([0,1,0,0]).astype(np.float)
          elif(target[pic_input_r] == 3):
                target_array = np.array([0,0,1,0]).astype(np.float)
          elif(target[pic_input_r] == 4):
                target_array = np.array([0,0,0,1]).astype(np.float)

          valid_round = 0
          #ถ้าไม่ถูกต้อง อัปเดต
          while not(np.array_equal(target_array,transfer_type)):
              valid_round += 1
              if valid_round == 1000:
                  valid_round =0
                  break
              WLT_R,b_for_WLT_R,delta_weight = tr_L2T(target[pic_input_r],transfer_type,WLT_R,b_for_WLT_R,in_r)
              delta_weight_trans =  np.transpose(delta_weight) 
              
              WIL_R,b_for_WIL_R  = tr_L(delta_weight_trans[0],WIL_R,b_for_WIL_R,in_r,0,in_r*3)
              WIL_R,b_for_WIL_R  = tr_L(delta_weight_trans[1],WIL_R,b_for_WIL_R,in_r,1,in_r*3+1)
              WIL_R,b_for_WIL_R  = tr_L(delta_weight_trans[2],WIL_R,b_for_WIL_R,in_r,2,in_r*3+2)

              percept_L1 =  perceptron(np.array([bins_order[in_r],input_r[pic_input_r][in_r]]),WIL_R[in_r][0],b_for_WIL_R[in_r*3]) #in_r*3 มาจากวนให้ถึง 96
              transfer_L1 =  tf_L1(percept_L1)

              #L2
              percept_L2 =  perceptron(np.array([bins_order[in_r],input_r[pic_input_r][in_r]]),WIL_R[in_r][1],b_for_WIL_R[in_r*3+1])
              transfer_L2 =  tf_L2(percept_L2)

              #L3
              percept_L3 =  perceptron(np.array([bins_order[in_r],input_r[pic_input_r][in_r]]),WIL_R[in_r][2],b_for_WIL_R[in_r*3+2])
              transfer_L3 =  tf_L1(percept_L3)

              #type
              percept_type  = perceptron(np.array([transfer_L1,transfer_L2,transfer_L3]),WLT_R[in_r],b_for_WLT_R[in_r])
              transfer_type  = tf_L2T(percept_type)
              print("result for type = ",transfer_type,"   real result = ",target_array)
          

          # break
        #break
    #break
        ###################



```
